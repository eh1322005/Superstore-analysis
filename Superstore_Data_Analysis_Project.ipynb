```python
#Fillna .........fill evey missing value in every column with "the mode"
for col in data.columns:
    mode = data[col].mode()[0]
    data[col] = data[col].fillna(mode)
```

```python
#drop missing values "if their count >=50%"
for i in data.columns:
  if (data[i].isna().sum() / data[i].isna().count())*100>=50:
    data.drop(i,axis=1,inplace=True)
```

# **(16)Visualization**

```python
#Graph(1):
#[Discount vs Profit]...Knowing if the discounts negatively affect the profits.
sns.scatterplot(data=data, x='Discount', y='Profit', alpha=0.5, color='black')
plt.title('Discount vs Profit')
plt.grid(True)
plt.show()
```

Insights of graph(1) :High discounts lead to lower profits



Recommendation for graph(1): Avoid giving too many discounts to protect profit



```python
#Graph(2):
#[Shipping Time]
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
data['Shipping Duration'] = (data['Ship Date'] - data['Order Date']).dt.days
data['Shipping Duration'].value_counts().sort_index().plot(kind='bar', color='green')
plt.title('Shipping Duration Distribution')
plt.xlabel('Shipping Days')
plt.ylabel('Number of Orders')
plt.grid(axis='y')
plt.show()

```


Insights of graph(2): Most orders are delivered in 4 to 5 days.


Recommendation for graph(2): Inform customers to expect delivery within 4–5 days

```python
#Graph(3):
#[Loss Making Products]
loss_data = data[data['Profit'] < 0]
loss_data['Product Name'].value_counts().head(5).plot(kind='bar', color='red')
plt.title('Top 5 Loss-Making Products')
plt.ylabel('Frequency')
plt.grid(axis='y')
plt.show()
```

Insights of graph(3):Some products frequently lose money

Recommendation oof graph(3):
Review the pricing or promotion strategy of these items.



```python
#Graph(4):
#[High Profit Margin Products]
data['Profit Margin'] = data['Profit'] / data['Sales']
top_margin = data.groupby('Product Name')['Profit Margin'].mean().sort_values(ascending=False).head(5)
top_margin.plot(kind='bar', color='gold')
plt.title('Top 5 Products by Avg Profit Margin')
plt.ylabel('Avg Profit Margin')
plt.grid(axis='y')
plt.show()
```

Insight of graph(4):Some products have a high profit margin even if they don’t sell a lot.

Recommendation for graph(4):Promote these high-margin products more to increase overall profit.



```python
#Graph(5):
#[Segment Performance""sales and profit""]
data.groupby('Segment')[['Sales', 'Profit']].mean().plot(kind='bar', colormap='Set2')
plt.title('Average Sales and Profit per Segment')
plt.ylabel('Amount ($)')
plt.grid(axis='y')
plt.show()
```

Insight of graph(5):Corporate and Home Office segments perform better

Recommendation for graph(5):Focus marketing efforts on Corporate and Home Office customer

```python
#Graph(6):
#[ Profit by Shipping Mode]
data.groupby('Ship Mode')['Profit'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Profit by Shipping Mode')
plt.ylabel('Average Profit')
plt.grid(axis='y')
plt.show()
```

Insight of graph(6):Second Class shipping shows higher average profit

Recommendation for graph(6):Encourage the use of Second Class shipping to boost profit.



```python
#Graph(7):
#[Profit from Top-Selling Products]
top_products = data['Product Name'].value_counts().head(5)
profit_from_top = data.groupby('Product Name')['Profit'].sum().loc[top_products.index]
profit_from_top.plot(kind='bar', color='darkorange')
plt.title('Profit for Top-Selling Products')
plt.ylabel('Total Profit')
plt.grid(axis='y')
plt.show()
```


Insight of graph(7):Insight:
Best-selling products are not always the most profitable


Recommendation for graph(7):Analyze both sales and profit before making stocking decisions.



```python
#Graph(8):
#Pareto Analysis (80/20 Rule)
#This code shows which products make up most of the orders, using the 80/20 rule (Pareto Analysis).
## The 80/20 rule says that about 80% of sales usually come from just 20% of the products.

import matplotlib.pyplot as plt
product_orders = data['Product Name'].value_counts()
cumulative_percentage = product_orders.cumsum() / product_orders.sum()
plt.figure(figsize=(10, 5))
cumulative_percentage.plot(drawstyle='steps-post', color='green')
plt.axhline(y=0.8, color='red', linestyle='--', label='80% Line')

plt.title('Pareto Analysis (Simple View)')
plt.ylabel('Cumulative % of Orders')
plt.xlabel('Products (sorted)')
plt.legend()
plt.grid(True)
plt.show()
```

 Insight of graph(8):Roughly 20% of products generate 80% of orders – Pareto principle

Recommendation for graph(8): Focus on selling and promoting these few products that people buy the most.


```python
#Graph(9):
#[Sales & Profit by Category and Sub-Category]
#----------------------------------------------------
# Sales & Profit by Category
category_sales = data.groupby("Category")[["Sales", "Profit"]].sum().sort_values("Sales", ascending=False)
category_sales

# Plot sales by category
plt.figure(figsize=(8,5))
sns.barplot(x=category_sales.index, y=category_sales["Sales"], palette="Blues_d")
plt.title("Total Sales by Category")
plt.ylabel("Sales")
plt.show()

# Sales & Profit by Sub-Category
subcategory_sales = data.groupby("Sub-Category")[["Sales", "Profit"]].sum().sort_values("Sales", ascending=False)
subcategory_sales

plt.figure(figsize=(12,6))
sns.barplot(x=subcategory_sales.index, y=subcategory_sales["Sales"], palette="viridis")
plt.title("Total Sales by Sub-Category")
plt.xticks(rotation=45)
plt.ylabel("Sales")
plt.show()
```

Insights and recommendatins for graph(9):

**Category Insights**:
Technology has highest sales and profit, Furniture has lowest margin....**Category Recommendation**:
Focus on Technology and improve pricing for Furniture.




**Sub_Category Insights**:
Phones and Chairs sell most, but Tables give low profit.....**Sub-Category Recommendation**:
Promote high-margin items and review low-profit pricing.



```python
#Graph(10):
#[Regional Performance Analysis]
#------------------------------------------------------------
# Sales & Profit by Region
region_sales = data.groupby("Region")[["Sales", "Profit"]].sum().sort_values("Sales", ascending=False)
region_sales

plt.figure(figsize=(7,5))
sns.barplot(x=region_sales.index, y=region_sales["Sales"], palette="coolwarm")
plt.title("Sales by Region")
plt.ylabel("Sales")
plt.show()

# Top 10 States by Sales
state_sales = data.groupby("State")[["Sales", "Profit"]].sum().sort_values("Sales", ascending=False).head(10)
state_sales

plt.figure(figsize=(12,6))
sns.barplot(x=state_sales.index, y=state_sales["Sales"], palette="magma")
plt.title("Top 10 States by Sales")
plt.xticks(rotation=45)
plt.ylabel("Sales")
plt.show()
```

Insights and recommendation for graph(10): **Regional Insights**:
West region has highest sales and profit, Central is lowet.....**Regional Recommendation**:
Focus marketing on West and improve strategy in Central.



```python
#Graph(11):
# Customers and Products
#--------------------------------------------------------------
# Top 10 Customers by Sales
top_customers = data.groupby("Customer Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_customers

plt.figure(figsize=(12,6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette="plasma")
plt.title("Top 10 Customers by Sales")
plt.xlabel("Sales")
plt.show()

# Top 10 Products by Sales
top_products = data.groupby("Product Name")["Sales"].sum().sort_values(ascending=False).head(10)
top_products

plt.figure(figsize=(12,6))
sns.barplot(x=top_products.values, y=top_products.index, palette="cividis")
plt.title("Top 10 Products by Sales")
plt.xlabel("Sales")
plt.show()
```

Insights and recommendation for graph(11): **Customer Insights**:
Few loyal customers generate the highest sales......**Customer Recommendation**:
Keep strong relationships with top customers and offer loyalty rewards.





**Product Insights**:
A small number of products drive most of the sales......**Product Recommendation**:
Focus inventory and promotions on top-selling products.
